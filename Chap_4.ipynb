```python
import pandas as pd

url = "https://raw.githubusercontent.com/amueller/mglearn/main/mglearn/data/adult.data"

col_names = ['age', 'workclass', 'fnlwgt', 'education', 'edu-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'cap-loss', 'hours-per-week', 'native-country', 'income']

adult_data = pd.read_csv(url, names = col_names)


#IPython.display gives nice formatting to Jupyter notebooks
adult_data.head() #displays all the columns

```

```python
#Select some cols
data = adult_data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]

display(data.head())
```

```python
#value_counts method checks col contents
print(data.gender.value_counts()) #will show male 21790, female 10771
```

```python
print("Original features: \n", list(data.columns), " \n")
data_dummies = pd.get_dummies(data)
print("Features after get_dummies: \n", list(data_dummies.columns))
```

```python
#display 1st 5 rows
data_dummies.head()
```

```python
features = data_dummies.loc[:, 'age':'occupation_ Transport-moving']
#Extract NumPy arrays
x = features.values
y = data_dummies['income_ >50K'].values
print("x.shape: {} y.shape: {}".format(x.shape, y.shape))
```
