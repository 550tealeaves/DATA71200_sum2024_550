```python
import pandas as pd

url = "https://raw.githubusercontent.com/amueller/mglearn/main/mglearn/data/adult.data"

col_names = ['age', 'workclass', 'fnlwgt', 'education', 'edu-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'cap-loss', 'hours-per-week', 'native-country', 'income']

adult_data = pd.read_csv(url, names = col_names)


#IPython.display gives nice formatting to Jupyter notebooks
adult_data.head() #displays all the columns

```

```python
#Select some cols
data = adult_data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]

display(data.head())
```

```python
#value_counts method checks col contents
print(data.gender.value_counts()) #will show male 21790, female 10771
```

```python
#Use get_dummies function to transform all columns that have object type(ex: strings) or are categorical

print("Original features: \n", list(data.columns), " \n")
data_dummies = pd.get_dummies(data)
print("Features after get_dummies: \n", list(data_dummies.columns))
```

```python
#display 1st 5 rows - shows that orig cat feat expanded into 1 new feat for each possible value
#example says that these are supposed to have 0.0, or 1.0
data_dummies.head()
```

```python
#Use values attribute to convert data_dummies DataFrame into NumPy array & then train mlm
#Make sure to separate target var from data prior to training
#Common mistake is to include output var or some derived property of output var

features = data_dummies.loc[:, 'age':'occupation_ Transport-moving']
#Extract NumPy arrays
x = features.values
y = data_dummies['income_ >50K'].values
print("x.shape: {} y.shape: {}".format(x.shape, y.shape))
```

```python
#data now ready for scikit-learn to work w/

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#Split data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
print("Test score: {:.2f}".format(logreg.score(x_test, y_test)))
```
