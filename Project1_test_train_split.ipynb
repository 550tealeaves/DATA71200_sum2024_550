```python
import os
import tarfile
from six.moves import urllib


DOWNLOAD_ROOT = "https://raw.githubusercontent.com/550tealeaves/DATA71200_sum2024_550/main/project%201/pizza_v2.csv"
pizza_PATH = os.path.join("datasets", "pizza")
pizza_URL = DOWNLOAD_ROOT + "datasets/pizza/pizza_v2.tgz"
#pizza_URL = DOWNLOAD_ROOT + "project 1/pizza_v2.tgz"

def fetch_pizza_data(pizza_url=pizza_URL, pizza_path=pizza_PATH):
    if not os.path.isdir(pizza_path):
        os.makedirs(pizza_path)
        tgz_path = os.path.join(pizza_path, "pizza.tgz")
        urllib.request.urlretrieve(pizza_url, tgz_path)
        pizza_tgz = tarfile.open(tgz_path)
        pizza_tgz.extractall(path=pizza_path)
        pizza_tgz.close()
```
